```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 파일 경로 설정
file_path = 'C:/Users/winbb/Desktop/final.xlsx'

# 엑셀 파일 읽기
df = pd.read_excel(file_path)

# '합계'가 포함된 행 제외
df_filtered = df[~df['연령별'].str.contains('합계')]

# 단순선형회귀분석을 위한 데이터 선택
X = df_filtered['인구 (명)']  # 독립 변수 선택
Y = df_filtered['구성비 (%)']  # 종속 변수 선택

# 상수 항 추가
X = sm.add_constant(X)

# 단순선형회귀모델 학습
model = sm.OLS(Y, X)
results = model.fit()

# 회귀분석 결과 출력
print(results.summary())

# 회귀선과 산점도 그래프 그리기
plt.scatter(X['인구 (명)'], Y, alpha=0.5)  # 산점도
plt.plot(X['인구 (명)'], results.predict(), color='red')  # 회귀선
plt.xlabel('인구 (명)')
plt.ylabel('구성비 (%)')
plt.title('단순선형회귀 분포도')
plt.show()
```
